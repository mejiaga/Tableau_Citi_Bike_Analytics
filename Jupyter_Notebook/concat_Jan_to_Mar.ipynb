Concatenate the dataframes of each month and remove any row that has NaN values, then export the 1st Quater 2018 dataframe as csv.

```python
df_all =  pd.concat([df6, df7, df8, df9, df10, df11, df12], ignore_index=True)
df_all.tail()
```

```python
df_all['gender']
```

```python
df_all['Gender'] = df_all['gender'].replace([0,1,2],['unknown',"Male","Female"])
df_all = df_all.drop('gender',1)
df_all["distance"] = ''
df_all.tail()
```

```python
import geopy.distance
for index, row in df_all.iterrows():
    coords_1 = (row['start station latitude'],row['start station longitude'])
    coords_2 = (row['end station latitude'],row['end station longitude'])
    df_all.set_value(index,'distance',geopy.distance.vincenty(coords_1, coords_2).km)
df_all.tail()
```

```python
df_all.isnull().any() # Null values in column "Birth Year" detected
```

```python
df_all.to_csv('../Data/17_June_to_Dec_citibikenyc_tripdata.csv', index_label = "TripID")
```
